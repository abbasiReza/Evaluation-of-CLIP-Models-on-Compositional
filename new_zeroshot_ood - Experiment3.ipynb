```python
import open_clip
import torchvision
import torch
import torch.nn.functional as F

all_models=open_clip.list_pretrained()
```

```python
def accuracy(output, target, topk=(1,)):
    pred = output.topk(max(topk), 1, True, True)[1].t()
    # print(pred)
    # print(target)

    correct = pred.eq(target.view(1, -1).expand_as(pred))
    # print(correct)
    # print([float(correct[:k].reshape(-1).float().sum(0, keepdim=True).cpu().numpy()) for k in topk])
    return [float(correct[:k].reshape(-1).float().sum(0, keepdim=True).cpu().numpy()) for k in topk]
```

```python
import pandas as pd
df_ood=pd.DataFrame(columns=['model','ood_dis_Acc_top1','ood_dis_Acc_top5'])
df_ood
```

```python

autocast = torch.cuda.amp.autocast
cast_dtype = None
for i,a in enumerate(all_models):
    try:
        print(a)
        model, _, preprocess = open_clip.create_model_and_transforms(a[0], pretrained=a[1])
        tokenizer = open_clip.get_tokenizer(a[0])
        model.cuda()
        model.eval()
        dataset=torchvision.datasets.ImageFolder('C:/Users/user01/Documents/datasets/ood_only_names/',transform=preprocess)
        dataloader=torch.utils.data.DataLoader(dataset,shuffle=False,batch_size=64
                                        )
        


        # Example list of texts to encode
        
        with torch.no_grad():
        # Specify batch size
            batch_size = 80

            # Encode texts in batches
            text_features_target = []
            for i in range(0, len(all_texts), batch_size):
                batch_texts = all_texts[i:i+batch_size]
                text = tokenizer(batch_texts)
                with torch.no_grad():
                    text_features_batch = F.normalize(model.encode_text(text.cuda()))
                    # text.cuda()
                    # print(text_features_batch.shape)
                    text_features_batch=text_features_batch.mean(dim=0)

                    # text_features_batch /= text_features_batch.norm(dim=-1, keepdim=True)
                    text_features_batch /= text_features_batch.norm(dim=-1, keepdim=True)

                    # print(text_features_batch.unsqueeze(0).shape)
                    # print(text_features_batch.mean(dim=0).shape)
                    text_features_target.append(text_features_batch.unsqueeze(0))

            # Concatenate text features from batches
            text_features_target = torch.cat(text_features_target, dim=0)
        # print(a)    
        with torch.no_grad():
            top1, top5, n = 0., 0., 0.
            for images, target in dataloader:
                images = images.to('cuda')
                if cast_dtype is not None:
                    images = images.to(dtype=cast_dtype)
                target = target.to('cuda')


                with autocast():
                    # predict
                    image_features = model.encode_image(images)
                    # image_features = F.normalize(image_features, dim=-1)
                    logits = 100. * image_features @ text_features_target.T
                # print(image_features.shape)
                # measure accuracy
                # print(logits)
                acc1, acc5 = accuracy(logits, target, topk=(1, 5))
                # print(acc1, acc5)
                top1 += acc1
                top5 += acc5
                n += images.size(0)
            
        top1 = (top1 / n)
        top5 = (top5 / n)
        row={'model':a[0]+'_'+a[1],'ood_dis_Acc_top1':top1,'ood_dis_Acc_top5':top5}
        print(row)
        df_ood = df_ood.append(row, ignore_index=True)
        df_ood.to_csv('ood_50k_only_name.csv', index=False)

    except:
        row={'model':a[0]+'_'+a[1],'ood_dis_Acc_top1':0,'ood_dis_Acc_top5':0}
        print(row)
        df_ood = df_ood.append(row, ignore_index=True)
        continue
```

```python
import os
import shutil

def copy_files(source_folder, destination_folder):
    # Get the list of folders in the source folder
    folders = os.listdir(source_folder)

    for folder_name in folders:
        # Construct the source folder path
        source_folder_path = os.path.join(source_folder, folder_name)
        temp_folder = folder_name[folder_name.find(' ')+1:]

        if os.path.isdir(source_folder_path):
            # Construct the destination folder path
            destination_folder_path = os.path.join(destination_folder, temp_folder)

            if not os.path.exists(destination_folder_path):
                # Create the destination folder if it doesn't exist
                os.makedirs(destination_folder_path)

            # Get the list of files in the source folder
            files = os.listdir(source_folder_path)

            for file_name in files:
                # Construct the source file path
                source_file_path = os.path.join(source_folder_path, file_name)

                if os.path.isfile(source_file_path):
                    # Construct the destination file path
                    destination_file_path = os.path.join(destination_folder_path, file_name)

                    # Copy the file to the destination folder
                    shutil.copy(source_file_path, destination_file_path)

# Example usage
source_folder = 'C:/Users/user01/Documents/datasets/final_dataset'
destination_folder = 'C:/Users/user01/Documents/datasets/ood_only_names/'

copy_files(source_folder, destination_folder)
```
